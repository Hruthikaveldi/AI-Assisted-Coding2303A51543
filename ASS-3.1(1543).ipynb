> **Zero-Shot Prompt:**

**Write a Python function `is_palindrome_number(num)` that takes an integer `num` as input and returns `True` if it's a palindrome number, `False` otherwise. A palindrome number reads the same forwards and backward. Consider negative numbers and single-digit numbers.**



```python
def is_palindrome_number(num):
    if num < 0:
        return False
    s = str(num)
    return s == s[::-1]
```

Write a one-shot prompt by providing one input-output example and
ask the AI to generate a Python function to compute the factorial of a given number.

Example:

Input: 5 → Output: 120

Task:

>  **One-Shot Prompt:**

**Input: 5 → Output: 120**

**Write a Python function `calculate_factorial(num)` that takes an integer `num` as input and returns its factorial. The factorial of a non-negative integer `n` is the product of all positive integers less than or equal to `n`. The factorial of 0 is 1.**



```python
def calculate_factorial(num):
    if num < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    elif num == 0:
        return 1
    else:
        factorial = 1
        for i in range(1, num + 1):
            factorial *= i
        return factorial
```

# **Question 3: Few-Shot Prompting (Armstrong Number Check)**

Write a few-shot prompt by providing multiple input-output examples to guide the AI in generating a Python function to check whether a given number is an Armstrong number.

Examples:

• Input: 153 → Output: Armstrong Number

• Input: 370 → Output: Armstrong Number

• Input: 123 → Output: Not an Armstrong Number

Task:

• Analyze how multiple examples influence code structure and
accuracy.

• Test the function with boundary values and invalid inputs.

> Write a Python function is_armstrong_number(num) that checks if a given number is an Armstrong number, following this definition: A number is an Armstrong number if it is equal to the sum of its digits raised to the power of the number of digits.

Examples:

Input: 153 → Output: Armstrong Number

Input: 370 → Output: Armstrong Number

Input: 123 → Output: Not an Armstrong Number

Input: 9474 → Output: Armstrong Number

Input: 9475 → Output: Not an Armstrong Number

Test Cases:
Implement test cases for the following scenarios:

The provided examples.

Boundary values such as single-digit numbers (0, 1, 9).

Negative numbers and non-integer inputs (e.g., strings, floats).



```python
def is_armstrong_number(num):
    # Convert the number to a string to get its digits and count them
    num_str = str(num)
    num_digits = len(num_str)

    # Calculate the sum of its own digits raised to the power of the number of digits
    sum_of_powers = 0
    for digit in num_str:
        sum_of_powers += int(digit) ** num_digits

    # Check if the number is equal to the sum of its own digits raised to the power of the number of digits
    if sum_of_powers == num:
        return "Armstrong Number"
    else:
        return "Not an Armstrong Number"
```

```python
def test_is_armstrong_number(func):
    test_cases = [
        # Provided examples
        (153, "Armstrong Number"),
        (370, "Armstrong Number"),
        (123, "Not an Armstrong Number"),
        (9474, "Armstrong Number"),
        (9475, "Not an Armstrong Number"),

        # Boundary values: single-digit numbers
        (0, "Armstrong Number"),  # 0^1 = 0
        (1, "Armstrong Number"),  # 1^1 = 1
        (9, "Armstrong Number"),  # 9^1 = 9

        # Other valid Armstrong numbers
        (371, "Armstrong Number"),
        (407, "Armstrong Number"),
        (1634, "Armstrong Number"),
        (8208, "Armstrong Number"),
        (9474, "Armstrong Number"),

        # Other non-Armstrong numbers
        (10, "Not an Armstrong Number"),
        (100, "Not an Armstrong Number"),
        (1000, "Not an Armstrong Number"),

        # Invalid inputs
        (-153, ValueError),  # Negative number
        (3.14, TypeError),   # Float
        ("abc", TypeError), # String
        (None, TypeError)    # None
    ]

    print("\nTesting is_armstrong_number function:")
    all_passed = True
    for num_input, expected_output in test_cases:
        try:
            result = func(num_input)
            if result == expected_output:
                print(f"  Input: {num_input}, Expected: {expected_output}, Got: {result} - PASSED")
            else:
                print(f"  Input: {num_input}, Expected: {expected_output}, Got: {result} - FAILED")
                all_passed = False
        except (ValueError, TypeError) as e:
            if expected_output == type(e):
                print(f"  Input: {num_input}, Expected: {expected_output.__name__}, Got: {e.__class__.__name__} - PASSED (Error handling)")
            else:
                print(f"  Input: {num_input}, Expected: {expected_output}, Got: an unexpected error: {e} - FAILED")
                all_passed = False
        except Exception as e:
            print(f"  Input: {num_input}, Got an unexpected exception: {e} - FAILED")
            all_passed = False

    if all_passed:
        print("All test cases for is_armstrong_number function passed!")
    else:
        print("Some test cases for is_armstrong_number function failed. Review the logic.")

test_is_armstrong_number(is_armstrong_number)
```

# **Question 4: Context-Managed Prompting (Optimized Number Classification)**

Design a context-managed prompt with clear instructions and
constraints to generate an optimized Python program that classifies a
number as prime, composite, or neither.

Task:

• Ensure proper input validation.

• Optimize the logic for efficiency.

• Compare the output with earlier prompting strategies.

-----


Context-Managed Prompt:

**Task Description:**

Write a Python function classify_number(num) that classifies a given number num as:

Prime if the number is greater than 1 and has no divisors other than 1 and itself.

Composite if the number is greater than 1 and has divisors other than 1 and itself.

Neither if the number is less than or equal to 1 (since 0 and 1 are neither prime nor composite).

**Requirements:**

Input Validation: The function should handle invalid inputs, ensuring that the input is a valid integer.

Optimization: Optimize the logic for efficiency, particularly for large numbers (e.g., by checking divisibility up to the square root of the number).

Edge Cases: Handle edge cases like 0, 1, negative numbers, and non-integer inputs.

Expected Output:

The function should return 'Prime', 'Composite', or 'Neither' based on the input.

**Test Cases:**

Input: 29 → Output: "Prime"

Input: 15 → Output: "Composite"

Input: 1 → Output: "Neither"

Input: -7 → Output: "Neither"

Input: 0 → Output: "Neither"

Input: "abc" → Output: "Invalid input" (for non-integer input)

```python
import math

def classify_number(num):
    # Input Validation
    if not isinstance(num, int):
        return "Invalid input"

    # Handle edge cases as Neither
    if num <= 1:
        return "Neither"

    # Check for Prime or Composite
    # Optimization: Check divisibility only up to the square root of the number
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return "Composite"

    return "Prime"
```

```python
def test_classify_number(func):
    test_cases = [
        (29, "Prime"),
        (15, "Composite"),
        (1, "Neither"),
        (-7, "Neither"),
        (0, "Neither"),
        ("abc", "Invalid input"), # Non-integer input
        (2, "Prime"),
        (4, "Composite"),
        (7, "Prime"),
        (101, "Prime"), # Larger prime
        (100, "Composite"), # Larger composite
        (3.14, "Invalid input") # Float input
    ]

    print("\nTesting classify_number function:")
    all_passed = True
    for num_input, expected_output in test_cases:
        result = func(num_input)
        if result == expected_output:
            print(f"  Input: {num_input}, Expected: {expected_output}, Got: {result} - PASSED")
        else:
            print(f"  Input: {num_input}, Expected: {expected_output}, Got: {result} - FAILED")
            all_passed = False

    if all_passed:
        print("All test cases for classify_number function passed!")
    else:
        print("Some test cases for classify_number function failed. Review the logic.")

test_classify_number(classify_number)
```

# **Question 5: Zero-Shot Prompting (Perfect Number Check)**

Write a zero-shot prompt (without providing any examples) to
generate a Python function that checks whether a given number is a
perfect number.

Task:

• Record the AI-generated code.

• Test the program with multiple inputs.

• Identify any missing conditions or inefficiencies in the logic.



> Write a Python function is_perfect_number(num) that checks if a given number num is a perfect number. A perfect number is a positive integer that is equal to the sum of its proper divisors (excluding the number itself). For example, 6 is a perfect number because its divisors (excluding 6) are 1, 2, and 3, and their sum is 6. If the number is perfect, the function should return True; otherwise, it should return False.



```python
def is_perfect_number(num):
    # Perfect numbers are defined for positive integers.
    # Handle non-integer or non-positive inputs.
    if not isinstance(num, int):
        return "Invalid input"
    if num <= 0:
        return "Invalid input"

    # Calculate the sum of proper divisors
    sum_divs = 0
    for i in range(1, num):
        if num % i == 0:
            sum_divs += i

    # Check if the sum of divisors equals the number itself
    if sum_divs == num:
        return True
    else:
        return False
```

```python
def test_is_perfect_number(func):
    test_cases = [
        (6, True),          # Example: 1+2+3 = 6
        (28, True),         # Next perfect number: 1+2+4+7+14 = 28
        (496, True),        # Next perfect number
        (8128, True),       # Next perfect number
        (1, False),         # Not perfect (sum of divisors is 0)
        (2, False),         # Not perfect (sum of divisors is 1)
        (15, False),        # Not perfect (1+3+5 = 9)
        (30, False),        # Not perfect
        (0, "Invalid input"),   # Edge case: non-positive
        (-6, "Invalid input"),  # Edge case: negative
        (3.14, "Invalid input"),# Edge case: non-integer
        ("text", "Invalid input") # Edge case: non-integer
    ]

    print("\nTesting is_perfect_number function:")
    all_passed = True
    for num_input, expected_output in test_cases:
        result = func(num_input)
        if result == expected_output:
            print(f"  Input: {num_input}, Expected: {expected_output}, Got: {result} - PASSED")
        else:
            print(f"  Input: {num_input}, Expected: {expected_output}, Got: {result} - FAILED")
            all_passed = False

    if all_passed:
        print("All test cases for is_perfect_number function passed!")
    else:
        print("Some test cases for is_perfect_number function failed. Review the logic.")

test_is_perfect_number(is_perfect_number)
```

Examples:

• Input: 8 → Output: Even

• Input: 15 → Output: Odd

• Input: 0 → Output: Even

Task:

• Analyze how examples improve input handling and output clarity.

• Test the program with negative numbers and non-integer inputs.




> Write a Python function is_even_or_odd(num) that takes an integer num as input and returns "Even" if the number is even, and "Odd" if the number is odd. The function should include input validation to ensure that the input is a valid integer. If the input is not a valid integer (e.g., a string, float, or None), the function should return "Invalid input". Additionally, the function should handle negative numbers properly, as they can also be classified as even or odd.

Examples:

Input: 8 → Output: Even

Input: 15 → Output: Odd

Input: 0 → Output: Even

Input: -2 → Output: Even

Input: -7 → Output: Odd

Input: "abc" → Output: Invalid input

Input: 3.14 → Output: Invalid input

Input: None → Output: Invalid input




```python
def is_even_or_odd(num):
    # Input validation: Check if the input is an integer
    if not isinstance(num, int):
        return "Invalid input"

    # Handle even and odd for both positive and negative integers
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
# Test cases
test_cases = [
    (8, "Even"),
    (15, "Odd"),
    (0, "Even"),
    (-2, "Even"),
    (-7, "Odd"),
    ("abc", "Invalid input"),
    (3.14, "Invalid input"),
    (None, "Invalid input")
]

print("Testing is_even_or_odd function:")
for input_val, expected_output in test_cases:
    result = is_even_or_odd(input_val)
    if result == expected_output:
        print(f"  Input: {input_val}, Expected: {expected_output}, Got: {result} - PASSED")
    else:
        print(f"  Input: {input_val}, Expected: {expected_output}, Got: {result} - FAILED")
```
